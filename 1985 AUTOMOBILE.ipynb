```python
# Calculate the average of the "normalized-losses" column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)

# Print the average of "normalized-losses"
print("Average of normalized-losses: ", avg_norm_loss)

# Calculate the average of the "bore" column
avg_bore = df['bore'].astype('float').mean(axis=0)

# Print the average of "bore"
print("Average of bore: ", avg_bore)

# Calculate the average of the "stroke" column
avg_stroke = df["stroke"].astype("float").mean(axis=0)

# Print the average of "stroke"
print("Average of stroke:", avg_stroke)

# Calculate the average of the "horsepower" column
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)

# Print the average of "horsepower"
print("Average horsepower:", avg_horsepower)

# Calculate the average of the "peak-rpm" column
avg_peakrpm = df['peak-rpm'].astype('float').mean(axis=0)

# Print the average of "peak-rpm"
print("Average peak rpm:", avg_peakrpm)
```


Replace "NaN" by mean value in columns

```python
# Replace missing values with the respective average values v1
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
df["bore"].replace(np.nan, avg_bore, inplace=True)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
# Replace missing values with the respective average values v2
# Create a list of columns to process
columns_to_fill = ["normalized-losses", "stroke", "bore", "horsepower", "peak-rpm"]

# Loop through each column and replace missing values with the average value of that column
for col in columns_to_fill:
# Calculate the mean value of the column   
    avg_value = df[col].astype(float).mean() 
# Replace NaN value with average value
    df[col].fillna(avg_value, inplace=True) 
```

Correct data format

```python
# Display the data types of each column in the dataframe
df.dtypes
```

---> From the displayed information, certain columns don't have the right data format. Numbers should typically be labeled as 'float' or 'int', while categorical strings should be marked as 'object'. For instance, 'bore' and 'stroke' represent engine details and should be in 'float' or 'int' format, but they're currently labeled as 'object'. To fix this, we'll need to convert each column to the correct data type using the "astype()" method

```python
# Convert the "bore" and "stroke" columns to float data type
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")

# Convert the "normalized-losses" column to integer data type
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")

# Convert the "price" column to float data type
df[["price"]] = df[["price"]].astype("float")

# Convert the "peak-rpm" column to float data type
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```
